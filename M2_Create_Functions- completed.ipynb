### 1. Python program to add two numbers

1. The code should take user input.
2. The user should be able to add two float numbers. 
2. Create a function that will take two arguments(or parameters).
3. The function should print the first number and the second number and the sum

Example: adding 5.55 and 8.90 gives 14.45

```python

num1 = float(input("Enter your first number:"))
num2 = float(input("Enter your second number:"))

def addup (num1, num2):
    sum = num1 + num2
    print('The sum of {0} and {1} is {2}'.format(num1, num2, sum))
    return sum

print(addup(num1,num2))
```

### 4. Program to find the area of a circle 

The area of a circle can simply be evaluated using the following formula.

$Area = pi * r^{2}$ where r is radius of circle 

```python
import math

r = float(input("Enter circle radius: " ))

def circle_area(r):
    area = math.pi*r*r
    return area

x= (circle_area(r))

print("Area:",("%.2f" % x))
```

### 6. Program to swap the first and last element in the list

```python
list1 = [10, 20, 4, 45, 99]

def swap(list1):
    while list1 == list1:
        list1.insert(0,list1[-1])
        list1.append(list1[1])
        list1.pop(-2)
        list1.pop(1)
        return list1

print(swap(list1))
```

### 7a. Program to count occurrences of all elements in a list

```python
lst = ['A', 'B', 'A', 'd', 'B', 'C', 'd','A', 'A', 5, 4, 'C', 'B', 'B', 60]

frequency = {}
   
for item in lst:                    # iterating over the list
    if item in frequency:           # checking the element in dictionary
            frequency[item] += 1    # incrementing the count
    else:
            frequency[item] = 1     # initializing the count

print(frequency)
```

### 7b. Program to count occurrences of an element in a list

```python
def count_oc(lst, x):
    count = 0
    for element in lst:
        if (element == x):
            count = count + 1
    return count
 
lst = [8, 6, 8, 10, 6, 7, 10, 8, 8, 5, 4, 7, 6, 6, 60]
x = 6
print('{} has occurred {} times'.format(x,count_oc(lst, x)))
```

### 8. Program to print duplicates from a list of integers

```python
my_nums = [1, 2, 3, 4, 5, 4, 6, 5, 6, 5, 4, 20, 23, 30, 20]

def dupe(n):
    duplicate = []
    for num in range(len(n)):
        step = num+1
        for rp in range(step, len(n)):
            if n[num] == n[rp] and n[num] not in duplicate:
                duplicate.append(n[num])
    return duplicate

print (dupe(my_nums))
```

### 9. Program to check if a string is palindrome

Given a string, write a python function to check if it is palindrome or not. A string is said to be a palindrome if the reverse of the string is the same as the string. 

Example palindromes 

* "level"
* "radar"
* "rotor"

```python
x= input("Enter a word:")

def palin(x):
    a = x[::-1]
    a = a.replace(' ', '')
    return a.lower()

a = palin(x)

if a==a:
        print(x +" = Palindrome")
else:
        print(x + " != Paliondrome")
```

### 10. Program to get the Words Frequency in String

Example imputs: 
"How many cans can a canner can, if a canner can can cans? A canner can can as many cans as a canner can, if a canner can can cans."

"How much wood would a woodchuck chuck If a woodchuck could chuck wood? He would chuck, he would, as much as he could, And chuck as much as a woodchuck would If a woodchuck could chuck wood."

"She sells seashells by the seashore. The shells she sells are surely seashells. So if she sells shells on the seashore, I'm sure she sells seashore shells."



```python
str = "How many cans can a canner can, if a canner can can cans? A canner can can as many cans as a canner can, if a canner can can cans."

def change(str):
    l1 = str.split(" ")
    return l1 

frequency = {}
   
for item in change(str):            
    if item in frequency:           
        frequency[item] += 1    
    else:
        frequency[item] = 1     

print(frequency)
```

### 11. Python extract Unique values dictionary values

input:
student_list = {'Andrew': [5, 6, 7, 8],
             'Stephen': [10, 11, 7, 5],
             'Catherine': [6, 12, 10, 8],
             'David': [1, 2, 5]}
             
             
output: [1, 2, 5, 6, 7, 8, 10, 11, 12]

```python
student_list = {'Andrew': [5, 6, 7, 8], 'Stephen': [10, 11, 7, 5], 'Catherine': [6, 12, 10, 8], 'David': [1, 2, 5]}

sList = list(student_list.values())

def flatten(sList):
    return [item for sublist in sList for item in sublist]
    
nums = flatten(sList)
nums.sort()
nums = list(dict.fromkeys(nums))
print(nums)
```

```python
student_list = {'Andrew': [5, 6, 7, 8], 'Stephen': [10, 11, 7, 5], 'Catherine': [6, 12, 10, 8], 'David': [1, 2, 5]}

dList = list(student_list.values())

outp = []
for sublist in dList:
    for item in sublist:
        outp.append(item)

outp.sort()
outp = list(dict.fromkeys(outp))

print(outp)
```
