```python
#Exercise: Create a class called "Triangle" that has a method for calculating the area of the triangle using the base and height. 
#Use a @staticmethod decorator to indicate that the method doesn't need access to any class or instance level attribute.

''' @staticmethod applied'''

class Triangle:
    def __init__(self, area):
        self.area = area
       
    def __repr__(self):
        return f'<Area of the Triangle = {self.area:.3f}>'

    @staticmethod                                 # no longer need to create an object before calling the method
    def Tarea(base, height):                      # so, self, is removed
        return Triangle((base * height)/2)
                                                   
Area = Triangle.Tarea(5, 4.5)                     # not needing to pass an object
print(Area)
```

```python
''' @classmethod applied'''

class Circle: 
    import math as M
    radius = float(input('Enter radius:' ))
    PI = M.pi

    @classmethod
    def Circumference(cls): 
        print('Circumference of the circle is:', round(2*(cls.PI * cls.radius), 2))

C1 = Circle()
C1.Circumference( )
```
